```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/content/photo.jpg")
```

```python
print("ORIGINl IMAGE",img)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# Invert the grayscale image
inverted_image = 255 - gray_image
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(inverted_image, cmap='gray')
plt.title('Inverted Image')
plt.axis('off')
plt.show()
```

```python
# Apply Gaussian blur to the inverted image
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')
plt.show()
```

```python
# Invert the blurred image
inverted_blurred_image = 255 - blurred_image
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(inverted_blurred_image , cmap='gray')
plt.title('inverted_blurred_image ')
plt.axis('off')
plt.show()
```

```python
# Create the pencil sketch by blending the grayscale and inverted blurred image
pencil_sketch = cv2.divide(gray_image, inverted_blurred_image, scale=256.0)
```

```python
# Display the pencil sketch
plt.figure(figsize=(8, 8))
plt.imshow(pencil_sketch, cmap='gray')
plt.title('Pencil Sketch')
plt.axis('off')
plt.show()
```
